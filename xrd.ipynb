```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction import xrd

# --------------------------------------------------------
# Procedure de telechargement du fichier .cif du LiCoO2.
# --------------------------------------------------------

with MPRester("L41ZHCirMyRASy8QHJEO") as m:
    
    structure = m.get_structure_by_material_id("mp-22526")
    
    myCifWritten = CifWriter(structure)
    
    myCifWritten.write_file('LiCoO2_mp-22526_symmetrized.cif')
    
# --------------------------------------------------------
# Calcul des pics du diffractogramme.
# --------------------------------------------------------
    
    XrayDiffrac = xrd.XRDCalculator(wavelength='CuKa',symprec=0, debye_waller_factors=None)
    
    PatternXray = XrayDiffrac.get_pattern(structure, scaled=True, two_theta_range=(0,90))
    
    Indices_hkl = PatternXray.hkls ; Angles = PatternXray.x ; Intensities = PatternXray.y

# --------------------------------------------------------
# Affichage et résultats.
# --------------------------------------------------------
    
    print("--------------------------------------------------------")
    
    print("Determination des 3 premiers pics du diffractogramme correspondant") 
    print("pour un rayonnement X incident Cu Kα (\u03BB = 1.54060 \u212B)")
    print("avec affichage des indices hkl associes.\n")
    
    print("--------------------------------------------------------")
    
    print("Affichage de l'entièreté des pics du diffractogramme correspondant.\n")
    
    XrayDiffrac.get_plot(structure,two_theta_range=(0,90),annotate_peaks=True, ax=None,with_labels=True)
    
    print("--------------------------------------------------------")
    
    print(" -Premier Pic- \n")
    print(" > Angle : 2\u03B8   = " + str(Angles[0]))
    print(" > Intensite         = " + str(Intensities[0]))
    print(" > Indices de Miller = " + str(Indices_hkl[0][0]['hkl']))
    print(" _____________________\n")
    
    print(" -Premier Pic- \n")
    print(" > Angle : 2\u03B8   = " + str(Angles[1]))
    print(" > Intensite         = " + str(Intensities[1]))
    print(" > Indices de Miller = " + str(Indices_hkl[1][0]['hkl']))
    print(" _____________________\n")
    
    print(" -Troisieme Pic- \n")
    print(" > Angle : 2\u03B8   = " + str(Angles[2]))
    print(" > Intensite         = " + str(Intensities[2]))
    print(" > Indices de Miller = " + str(Indices_hkl[2][0]['hkl']))
    print(" _____________________\n")
    
    print("--------------------------------------------------------")
    
# --------------------------------------------------------
```
